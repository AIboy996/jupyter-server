```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import scipy.optimize as optim
from collections import OrderedDict
```

# 6.2

## 写函数

```python
f = lambda x:pow(2*np.pi,-.5)*np.exp(-x**2/2)
```

```python
logf = lambda x:-.5*np.log(2*np.pi)-x**2/2
```

$$x_i=\frac{\log\frac{f\left(c_{i+1} \ \right)}{f(c_i)}+c_{i+1}^2-c^2_i}{c_{i+1-c_i}}$$

```python
def goal(c):
    '''envelope的面积减去logf的面积'''
    # 切线
    y = (lambda x: -c_i * (x - c_i) + logf(c_i) for c_i in c)
    # 积分区间 0-np.inf
    x_i = np.append(0, (np.log(f(c[1:]) / f(c[:-1])) + c[1:]**2 - c[:-1]**2) /
                    (c[1:] - c[:-1]))
    x_i = np.append(x_i, np.inf)
    result = np.sum([
        integrate.quad(lambda x: np.exp(y_i(x)) - f(x), x_i[i], x_i[i + 1])
        for i, y_i in enumerate(y)
    ])
    return result
```

```python
goal(np.linspace(1, 3, 3))
```

```python
def plot(c,xmax):
    '''绘制对数图像'''
    # 切线
    y = (lambda x: -c_i * (x - c_i) + logf(c_i) for c_i in c)  #这里必须用生成器
    x = np.linspace(1e-4, xmax, 1000)
    x_i = np.append(0, (np.log(f(c[1:]) / f(c[:-1])) + c[1:]**2 - c[:-1]**2) /
                    (c[1:] - c[:-1]))
    x_i = np.append(x_i, xmax)
    plt.plot(x, logf(x),c='k',label = r'$\log f(x)$')
    for i,y_i in enumerate(y):
        x_y_i = np.linspace(x_i[i],x_i[i+1],500)
        plt.scatter(c[i],logf(c[i]),c='red',label='tangent')
        plt.plot(x_y_i, y_i(x_y_i),c='blue',linestyle='--',label='Envelope')
    # 不重复显示图例
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())
```

```python
def plot_exp(c,xmax):
    '''绘制原图像'''
    # 切线
    y = (lambda x: -c_i * (x - c_i) + logf(c_i) for c_i in c)  #这里必须用生成器
    x = np.linspace(1e-4, xmax, 1000)
    x_i = np.append(0, (np.log(f(c[1:]) / f(c[:-1])) + c[1:]**2 - c[:-1]**2) /
                    (c[1:] - c[:-1]))
    x_i = np.append(x_i, xmax)
    plt.plot(x, np.exp(logf(x)),c='k',label='$f(x)$')
    for i,y_i in enumerate(y):
        x_y_i = np.linspace(x_i[i],x_i[i+1],500)
        plt.scatter(c[i],np.exp(logf(c[i])),c='red',label='tangent')
        plt.plot(x_y_i, np.exp(y_i(x_y_i)),c='blue',linestyle='--',label='Envelope')
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())
```

```python
plot(np.array([1,2]),3)
```

```python
plot_exp(np.array([1,2]),3)
```

## 最优化(n=1

```python
n = 1
# 观察随着切点变化函数值变化
l = []
for i in np.linspace(.01,3,5000):
    l.append(goal(np.array([i])))
plt.plot(np.linspace(.01,5,5000),l)
```

```python
# 遍历格点，求得最优值c=1
c = np.reshape(np.linspace(.01,3,5000)[l.index(min(l))],newshape=(1,))
c
```

```python
plot(c,3)
```

```python
plot_exp(c,3)
```

## 使用scipy.optimize包最优化

### n=1

```python
n = 1
solve = optim.minimize(goal,x0=np.linspace(1,n,n),method='CG')
plot(solve['x'],3)
solve
```

```python
plot_exp(solve['x'],3)
```
