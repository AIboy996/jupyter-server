```r
library(TSA)
```

```r
data(JJ)
```

# a 对数变换

```r
options(repr.plot.width=10, repr.plot.height=6)
plot(JJ,main='JJ')
```

```r
plot(log(JJ),main='log JJ')
```

**取对数之前的序列，有明显的非线性趋势；而取对数之后，转换为线性趋势，因此取对数变换是合理的\
实际上序列也存在明显的季节项**

# b 对数差分

```r
plot(diff(log(JJ)))
```

**对数差分之后的序列基本平稳，但是存在较为明显的异方差性，中间时段的方差较小**

# c 对数差分之后的ACF

```r
acf(diff(log(JJ)), ci.type='ma')
```

**对数差分之后的ACF呈现一定的周期下降趋势，lag=1、2、3的时候显著非0，又观察到序列的周期性，考虑做季节差分**

# d 季节差分之后的序列图

```r
plot(diff(diff(log(JJ)),lag=4))
```

**经过季节差分之后，前文提及的中间时段方差较小两边方差较大的问题得到解决，方差基本上随机分布**

# e 季节差分之后的ACF

```r
acf(diff(diff(log(JJ)),lag=4),ci.type='ma')
```

此时的ACF显著一阶截尾，在第二个Q3的时候，几乎显著不为0，基本可以认为这时候满足MA(1)模型

# f 拟合SARIMA(0,1,1)x(0,1,1)_4模型

```r
model = arima(log(JJ), order=c(0,1,1),seasonal = list(order=c(0,1,1),period=4))
```

```r
model
```

**ma1和sma1的系数都是显著的**

# g 模型诊断

## 残差满足正态性

```r
options(repr.plot.width=8, repr.plot.height=8)
qqnorm(residuals(model))
qqline(residuals(model))
```

## 残差满足同方差性、无自相关性

```r
options(repr.plot.width=10, repr.plot.height=18)
tsdiag(model)
```

LB检验的p值全部都很大，检验不显著，认为残差之间没有显著的自相关性\
残差的ACF也都显著为0\
学生化残差没有明显趋势，满足同方差性

# h 未来两年的预测
